```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_weather_model(a, b, c, time):
    return a * time**2 + b * time + c

# Example hardcoded coefficients for the quadratic equation and a time value
a = 0.01
b = 0.1
c = 15
time = np.linspace(0, 24, 100)  # Generating 100 time values from 0 to 24 hours

# Compute the weather parameter for each time value
weather_param = quadratic_weather_model(a, b, c, time)

# Plotting the quadratic weather model
plt.plot(time, weather_param, label='Weather Parameter')
plt.title('Quadratic Weather Model')
plt.xlabel('Time (hours)')
plt.ylabel('Weather Parameter')
plt.legend()
plt.grid(True)
plt.show()
```
